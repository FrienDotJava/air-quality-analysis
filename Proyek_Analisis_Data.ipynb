### Cleaning Data

#### Fill Missing Data

```python
dataframes = [aoti_df, changping_df, dingling_df, dongsi_df, guanyuan_df, gucheng_df, huairou_df, nongzhan_df, shunyi_df, tiantan_df, wanliu_df, wanshou_df]

for df in dataframes:
    #Fill numerical data
    cols_to_interpolate = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'WSPM']
    df[cols_to_interpolate] = df[cols_to_interpolate].interpolate(method='linear', limit_direction='forward')

    df['RAIN'] = df['RAIN'].ffill()

    #Fill categorical data
    df.wd = df.wd.fillna(value=df.wd.value_counts().keys()[0]) #Value diisi dengan value wd yang paling sering muncul di tiap database.

    
```

#### Add datetime columns

```python
for df in dataframes: 
    df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
    df['year_month'] = df['date'].dt.to_period('M')
    df['year_dt'] = df['date'].dt.to_period('Y')
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana rata-rata SO2, NO2 dan O3 pada tiap bulan di tahun tertentu dan distrik tertentu?

```python
monthly_avg_pollutant = aoti_df.groupby('year_month')[['SO2', 'NO2', 'O3']].mean().reset_index()
monthly_avg_pollutant['year_month'] = monthly_avg_pollutant['year_month'].dt.to_timestamp()

year = 2015
monthly_avg_pollutant = monthly_avg_pollutant[monthly_avg_pollutant['year_month'].dt.year==year]
monthly_avg_pollutant
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(monthly_avg_pollutant["year_month"], monthly_avg_pollutant["SO2"], label="SO2", marker='o')
ax.plot(monthly_avg_pollutant["year_month"], monthly_avg_pollutant["NO2"], label="NO2", marker='o')
ax.plot(monthly_avg_pollutant["year_month"], monthly_avg_pollutant["O3"], label="O3", marker='o')
plt.title("Average SO2, NO2, and O3 per Month (2015) in Aotizongxin")
plt.grid(True)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

### Pertanyaan 2: Distrik mana yang memiliki rata-rata kandungan PM2.5 paling tinggi dan paling rendah di tahun 2014-2016?

```python
fig, ax = plt.subplots(figsize=(11,8))

#year = 2014
#pm25_by_station_year = pm25_by_station_year[pm25_by_station_year['year_month'].dt.year==year]
df_station = pm25_by_station_year.groupby(by=['station'])

custom_colors = {'Dongsi': 'red', 'Dingling': 'blue'} 

for key, group in df_station:
    if key[0] in custom_colors:
        color = custom_colors[key[0]]
        opacity = 1
    else:
        color = 'gray'
        opacity = 0.4
    ax.plot(group['year_month'], group['avg_pm25'], label=key, color=color, alpha=opacity, marker="o")

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.xticks(rotation=45)
plt.title("Average PM2.5 per Month (2014-2016) in Every District")
plt.show()
```
